```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
n = 200
D = np.zeros((n, 2))

D[:,0] = np.linspace(0, 10, n) + np.random.randn(n)
D[:,1] = D[:, 0]**2 + np.random.randn(n)*10

# plt.plot(D[:,0], D[:, 1], "bo")
# plt.show()

sns.jointplot(D[:,0], D[:, 1])
```

```python
df = pd.DataFrame(data=D, columns=['var1', 'var2'])
sns.jointplot(df.columns[0], df.columns[1], data=df)
sns.jointplot(df.columns[0], df.columns[1], data=df, kind="kde" ) 
# sns.jointplot(data=df)
```

```python
x = np.linspace(-1, 1, n)
y1 = x**2
y2 = np.sin(3*x)
y3 = np.exp(-10*x**2)

sns.scatterplot(x=y1, y=y2, hue=y3, legend=False, palette='Spectral')
plt.show()
```

```python
# Exercise: create a regression plot of data in matrix D -> df

sns.regplot(df.columns[0], df.columns[1], data=df, color='green')
plt.title(f'Regression of {df.columns[0]} on {df.columns[1]}')
plt.show()
```
