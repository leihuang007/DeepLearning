```python
import matplotlib.pyplot as plt
import numpy as np
from imageio import imread
```

```python
# 矩阵可视化
m = 3
n = 5

M = np.random.randint(10, size=(m, n))
plt.imshow(M)

for i in range(m):
    for j in range(n):
        plt.text(j,i, str(M[i, j]), fontsize=20, horizontalalignment='center', verticalalignment='center') # image中的行列index和矩阵M中的正好相反。

plt.colorbar()
plt.show()
```
