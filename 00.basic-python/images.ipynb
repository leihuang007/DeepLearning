```python
# Exercise: Visualize the Hilbert matrix using seaborn (heatmap)
# H is 10x10
n = 10
H = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        H[i, j] = 1 / ((i+1)+(j+1)+1)
sns.heatmap(H, vmax=.4)
plt.show()
```
